### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for the plots
x_line = np.arange(0, 11, 1)
y_line = x_line ** 2

categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors_bar = ['blue', 'green', 'red', 'purple']

x_scatter = [1, 2, 3, 4, 5]
y_scatter = [2, 4, 6, 8, 10]

labels_pie = ['Python', 'Java', 'C++', 'JavaScript']
sizes_pie = [40, 25, 20, 15]
colors_pie = ['blue', 'green', 'red', 'yellow']

# Create the subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Plot a line chart in the first subplot
axs[0, 0].plot(x_line, y_line)
axs[0, 0].set_title('Line Chart: y = x^2')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True)

# Plot a bar chart in the second subplot
axs[0, 1].bar(categories, values, color=colors_bar)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Plot a scatter plot in the third subplot
axs[1, 0].scatter(x_scatter, y_scatter)
for i in range(len(x_scatter)):
    axs[1, 0].annotate(f'({x_scatter[i]}, {y_scatter[i]})', (x_scatter[i], y_scatter[i]), textcoords="offset points", xytext=(0,10), ha='center')
axs[1, 0].set_title('Scatter Plot with Annotations')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True)

# Plot a pie chart in the fourth subplot
axs[1, 1].pie(sizes_pie, labels=labels_pie, colors=colors_pie, autopct='%1.1f%%', startangle=140)
axs[1, 1].set_title('Pie Chart of Programming Languages')
axs[1, 1].axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
